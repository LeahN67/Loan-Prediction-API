```python
# Calculate missing values for each column in the train_loan dataset
missing_train = train_loan_data.isnull().sum().reset_index()
missing_train.columns = ['Column', 'Missing Values']
missing_train = missing_train[missing_train['Missing Values'] > 0]

print("Missing values in train dataset:\n", missing_train)
```

```python
# Calculate missing values for each column in the test_loan dataset
missing_test = test_loan_data.isnull().sum().reset_index()
missing_test.columns = ['Column', 'Missing Values']
missing_test = missing_test[missing_test['Missing Values'] > 0]

print("Missing values in test dataset:\n", missing_test)

```

```python
# Calculate missing values for each column in the repayment dataset
missing_repayment = repayment_data.isnull().sum().reset_index()
missing_repayment.columns = ['Column', 'Missing Values']
missing_repayment = missing_repayment[missing_repayment['Missing Values'] > 0]

print("Missing values in repayment dataset:\n", missing_repayment)
```
